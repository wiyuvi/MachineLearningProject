```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("onizlemeoncesi.csv")

trainingData = df.iloc[:, :].values
dataset = df.iloc[:, :].values
```

```python
df.head()
```

```python
df["RAM"] = df.RAM.str.replace(' GB', '')
df["Ekran_Boyutu"] = df.Ekran_Boyutu.str.replace(' İnç', '')
df["Cozunurluk"] = df.Cozunurluk.str.replace(' Piksel', '')
df["SSD"] = df.SSD.str.replace(' GB', '')
df["HDD"] = df.HDD.str.replace(' GB', '')
df["SSD"] = df.SSD.str.replace('1 TB', '1000')
df["HDD"] = df.HDD.str.replace('1 TB', '1000')
df["OS"] = df.OS.str.replace('\(İşletim Sistemi Yok\)', '')
df["price"] = df.price.str.replace('TL', '')
df["price"] = df.price.str.replace('+', '')
df["price"] = df.price.str.replace('Ücretsiz', '')
df["price"] = df.price.str.replace('Kargo', '')
df["price"] = df.price.str.replace(' ', '')
df.head()
```

```python
for i in range(10,100):
    df["price"] = df.price.str.replace(',{}'.format(i), '')

df["price"] = df.price.str.replace(',00', '')
df["price"] = df.price.str.replace(',01', '')
df["price"] = df.price.str.replace(',02', '')
df["price"] = df.price.str.replace(',03', '')
df["price"] = df.price.str.replace(',04', '')
df["price"] = df.price.str.replace(',05', '')
df["price"] = df.price.str.replace(',06', '')
df["price"] = df.price.str.replace(',07', '')
df["price"] = df.price.str.replace(',08', '')
df["price"] = df.price.str.replace(',09', '')
df["price"] = df.price.str.replace('.', '')
df.head()
```

```python
from sklearn.impute import SimpleImputer
imputerHDD = SimpleImputer(missing_values=np.nan, strategy='mean')
imputerHDD = imputerHDD.fit(trainingData[:, 4:5])
df["HDD"] = imputerHDD.transform(trainingData[:, 4:5])

imputerSSD = SimpleImputer(missing_values=np.nan, strategy='mean')
imputerSSD = imputerSSD.fit(trainingData[:, 5:6])
df["SSD"] = imputerSSD.transform(trainingData[:, 5:6])

imputerEKM = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imputerEKM = imputerEKM.fit(trainingData[:, 2:3])
df["Ekran_Karti_Modeli"] = imputerEKM.transform(trainingData[:, 2:3])
```

```python
df["RAM"] = df.RAM.astype(int)
df["HDD"] = df.HDD.astype(int)
df["SSD"] = df.SSD.astype(int)
df["Ekran_Boyutu"] = df.Ekran_Boyutu.astype(float)
df["price"] = df.price.astype(float)
```

```python
df.head(39)
```

```python
lineCounter = df["url"].count()

for i in range(0,lineCounter):
    df["Ekran_Karti_Modeli"].loc[i] = df["Ekran_Karti_Modeli"].loc[i].split()[0]
    df["Islemci"].loc[i] = df["Islemci"].loc[i].split()[0]
    df["OS"].loc[i] = df["OS"].loc[i].split()[0]
```

```python
df.head(12)
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse=False)

ohe.fit_transform(df[["Ekran_Karti_Modeli"]])
df["Ekran_Karti_Modeli"] = ohe.fit_transform(df[["Ekran_Karti_Modeli"]])

ohe.fit_transform(df[["Ekran_Karti_Modeli"]])
df["Ekran_Karti_Modeli"] = ohe.fit_transform(df[["Ekran_Karti_Modeli"]])

ohe.fit_transform(df[["Cozunurluk"]])
df["Cozunurluk"] = ohe.fit_transform(df[["Cozunurluk"]])

ohe.fit_transform(df[["Cozunurluk"]])
df["Cozunurluk"] = ohe.fit_transform(df[["Cozunurluk"]])

ohe.fit_transform(df[["Islemci"]])
df["Islemci"] = ohe.fit_transform(df[["Islemci"]])

ohe.fit_transform(df[["Islemci"]])
df["Islemci"] = ohe.fit_transform(df[["Islemci"]])

ohe.fit_transform(df[["OS"]])
df["OS"] = ohe.fit_transform(df[["OS"]])

ohe.fit_transform(df[["OS"]])
df["OS"] = ohe.fit_transform(df[["OS"]])

df.head(12)
```

```python
df.to_csv("onizlemesonrasi.csv", index=False)
```
