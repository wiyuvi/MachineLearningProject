```python
import pandas as pd 
import numpy as np
import requests #html kodlarını almak için
from bs4 import BeautifulSoup #veri çekmek için BeautifulSoup
from sklearn.preprocessing import LabelEncoder
```

```python
#Siteye tarayıcımızı tanıtıyoruz
headers_param={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.183 Safari/537.36 OPR/72.0.3815.320"}
```

```python
#Veri çekeceğimiz listeleri tanımlıyoruz
urlLaptop = []
pricesLaptop = []
propTitles = []
propValues = []

#26 ve 27. sayfadaki verileri çekiyoruz
for i in range(26,27):

    #ilgili url e bağlanıyoruz
    url = "https://www.epey.com/laptop/" + str(i)
    epey = requests.get(url, headers=headers_param)
   
    #bütün html içeriğni çekiyoruz
    htmlContent = epey.content

    #html kodlarını işleyebilmek için ayrıştırıyoruz
    parseHtml = BeautifulSoup(htmlContent, "html.parser")


    #işimize yarayacak html etiketlerini aratıyoruz
    allLaptops = parseHtml.find_all("div",{"class":"detay cell"})

    #laptopların linklerini urlLaptop a atıyoruz
    loop = 0
    for url in allLaptops:
        loop = loop+1
        #veri çekeceğim sitede ilk hücre reklam olduğu için bunu atlamasını istiyorum 
        if loop > 1:
            #mevcut indeksteki  laptopun linkini getLink'e atadık
            getLink = url.a.get("href") 
            print("Link: "+getLink)
            #linki diziye ekliyoruz
            urlLaptop.append(getLink)  
            
            #mevcut indeksteki laptopun linkine istek yolluyoruz
            currentLaptop = requests.get(getLink, headers=headers_param) 
            
            #mevcut indeksteki laptopun linkinin html içeriğini alıyoruz
            currentHtml = currentLaptop.content 

            #html içeriğini parse ediyoruz
            currentParse = BeautifulSoup(currentHtml, "html.parser")

            #laptopun fiyatını ekliyoruz
            laptopPrice = currentParse.find("span", {"class":"urun_fiyat"}).text
            pricesLaptop.append(laptopPrice)
            print("Fiyat: "+laptopPrice)
            #özelliklerini alacağım verilerin parent divine giriyoruz
            propertiesParent = currentParse.find("div", {"id":"oncelikli"})

            #id si boş olan classı cell olan divleri bul
            properties = propertiesParent.find_all("div", {"class":"cell","id":""})

            
            for propertie in properties:
                
                #Özelliğin başlığnı burda alıyorum
                propTitle = propertie.find("div", {"class":"row row1"})
               
                #Özelliğin değerini içeren divin classı değişken olabildiği için iki haliyle de çekiyorum
                #if koşullarıyla mevcut hücredeki özelliğin değerini içeren div hangi classa aitse
                #o classın değerini alıyorum
                propValue = propertie.find("div", {"class":"row row2"})
                propValue2 = propertie.find("div", {"class":"row row2 font12"})
                if propValue:
                    print("Başlık: "+propTitle.text)
                    propTitles.append(propTitle.text)
                    print("Özellik: "+propValue.text)
                    propValues.append(propValue.text)
                elif propValue2:
                    print("Başlık: "+propTitle.text)
                    propTitles.append(propTitle.text)
                    print("Özellik: "+propValue2.text)
                    propValues.append(propValue2.text)
```

```python
print(str(len(urlLaptop))+" adet link bulundu.")
print(str(len(pricesLaptop))+" adet fiyat bulundu.")
print(str(len(propTitles))+" adet başlık bulundu.")
print(str(len(propValues))+" adet değer bulundu.")       
```

```python
#dataframe oluşturuyruz
df_urls = pd.DataFrame()
df_urls["urls"] = urlLaptop
df_urls["prices"] = pricesLaptop

df_urls.head()
```

```python
laptops = len(urlLaptop)

#Bulunan propTitles leri kolon olarak yazdırıcaz. Bunun için unique e çevirip benzersizleri buluyoruz ki
#her özellik için kolon açılabilsin
columns = np.array(propTitles)
columns = np.unique(columns)

df = pd.DataFrame(columns=columns)
df["url"] = urlLaptop
df["price"] = pricesLaptop

df.head()
```

```python
for i in range(0, laptops):

    #kaçıncı indeksteki URL i almasını istedigimizi belirliyoruz
    url = df["url"].loc[i] 
    getData = requests.get(url)

    getContent = getData.content

    source = BeautifulSoup(getContent, "html.parser")

    propertiesParent = source.find("div", {"id":"oncelikli"})

            
    properties = propertiesParent.find_all("div", {"class":"cell","id":""})

            
    for propertie in properties:
                
        propTitle = propertie.find("div", {"class":"row row1"})
               
                
        propValue = propertie.find("div", {"class":"row row2"})
        propValue2 = propertie.find("div", {"class":"row row2 font12"})
        if propValue:
            print(propTitle.text+" "+propValue.text)
            df[propTitle.text].loc[i] = propValue.text
        elif propValue2:
            print(propTitle.text+" "+propValue2.text)
            df[propTitle.text].loc[i] = propValue2.text
```

```python
df.rename(columns={'Bellek (RAM):': 'RAM', 'Ekran Boyutu:': 'Ekran_Boyutu', 'Ekran Kartı Modeli:': 'Ekran_Karti_Modeli', 'Ekran Çözünürlüğü:': 'Cozunurluk', 'Sabit Disk (HDD) Boyutu:': 'HDD', 'Sabit Disk (SSD) Boyutu:': 'SSD', 'İşlemci Modeli:': 'Islemci', 'İşletim Sistemi:': 'OS'}, inplace=True)
```

```python
df["RAM"] = df.RAM.str.replace('\n', '')
df["Ekran_Boyutu"] = df.Ekran_Boyutu.str.replace('\n', '')
df["Ekran_Karti_Modeli"] = df.Ekran_Karti_Modeli.str.replace('\n', '')
df["Cozunurluk"] = df.Cozunurluk.str.replace('\n', '')
df["SSD"] = df.SSD.str.replace('\n', '')
df["Islemci"] = df.Islemci.str.replace('\n', '')
df["HDD"] = df.HDD.str.replace('\n', '')
df["OS"] = df.OS.str.replace('\n', '')
df["url"] = df.url.str.replace('\n', '')
df["price"] = df.price.str.replace('\n', '')
```
